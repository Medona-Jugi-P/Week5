```python
# Convert 'pcv', 'wc', 'rc' columns to float64
dataset['pcv'] = pd.to_numeric(dataset['pcv'], errors='coerce')
dataset['wc'] = pd.to_numeric(dataset['wc'], errors='coerce')
dataset['rc'] = pd.to_numeric(dataset['rc'], errors='coerce')
```

```python
dataset.describe()
```

```python
qual=[]
quan=[]

for columnNames in dataset.columns:
    print(columnNames)
    if(dataset[columnNames].dtype=='O'):
        print("qual")
        qual.append(columnNames)
    else:
        print("quan")
        quan.append(columnNames)
```

```python
quan
```

```python
qual
```

```python


# Numeric Columns
numeric_columns = ['age', 'bp', 'sg', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wc', 'rc']
dataset[numeric_columns] = dataset[numeric_columns].fillna(dataset[numeric_columns].median())

# Categorical Columns
categorical_columns = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane']
dataset[categorical_columns] = dataset[categorical_columns].fillna(dataset[categorical_columns].mode().iloc[0])

# Check if there are still any missing values
missing_values = dataset.isnull().sum()
print(missing_values)
```
