```python
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'Brown'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    # generate a sample
    sample = dataset
    # calculate parameters
    sample_mean =sample.mean()
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    # define the distribution
    dist = norm(sample_mean, sample_std)
    
    # sample probabilities for a range of outcomes
    values = [value for value in range(startrange, endrange)]
    probabilities = [dist.pdf(value) for value in values]    
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
```

```python
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot#library
    from scipy.stats import norm#library
    import seaborn as sns#library for visualization
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'Blue'},color='yellow')#kde=kernal density used to create curve in the plot
    pyplot.axvline(startrange,color='Red')#axvline = axis vertical line,starting range
    pyplot.axvline(endrange,color='Red')#ending range
    # generate a sample
    sample = dataset # sample datset
    # calculate parameters
    sample_mean =sample.mean() #finding mean and SD
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    # define the distribution
    dist = norm(sample_mean, sample_std)
    
    # sample probabilities for a range of outcomes
    values = [value for value in range(startrange, endrange)]#range convert into list by using this line of code
    probabilities = [dist.pdf(value) for value in values]    
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
```

```python
get_pdf_probability(dataset["ssc_p"],40,40)#same range will not be exist because there is no density in it.
```

# Standard Normal Distribution

```python
def stdNBgraph(dataset):
        # Coverted to standard Normal Distribution
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()

    values=[i for i in dataset]

    z_score=[((j-mean)/std) for j in values]

    sns.distplot(z_score,kde=True)

    sum(z_score)/len(z_score)
    #z_score.std()
```

```python
stdNBgraph(dataset["ssc_p"])
```
